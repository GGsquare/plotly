```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import numpy as np
```

```python
prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),
]
```

```python
df = pd.DataFrame(prices)
df = pd.DataFrame(prices, columns=["month", "price"])
df = df.set_index("month")
df.index
```

```python
#df['priceUSD'] = df['price']/4
df.insert(loc=0, column='priceUSD', value=df['price']/4)
df
```

```python
df.plot(title='Price of goods (USD)')
```

```python
plt.plot(df['priceUSD'], 'r--')
plt.suptitle('Price of goods(USD)')
plt.axis([0, 12.5, 0, 2])
plt.grid(True)
plt.xlabel('month')
```
